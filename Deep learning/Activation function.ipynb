# Activation Functions Quick Summary

## Problem Activation Functions Solve

Neural networks need to learn complex patterns such as:

* Images → Objects
* Text → Sentiment
* Features → Predictions

These relationships are usually **non-linear**, not straight lines.

---

## What Happens Without Activation Functions?

A neural network layer performs:

[
z = Wx + b
]

This is a linear transformation.

If we stack multiple layers without activations:

[
W_3(W_2(W_1x))
]

all layers collapse into a single linear transformation:

[
Ax + c
]

Therefore:

* 2 layers = linear model
* 20 layers = linear model
* 200 layers = linear model

The network gains no additional expressive power.

### Key Insight

Without activation functions, a deep neural network is mathematically equivalent to linear regression.

---

## Why Activation Functions Matter

Activation functions introduce non-linearity between layers.

Instead of:

[
Wx+b
]

we get:

[
Activation(Wx+b)
]

This prevents layers from collapsing into one linear transformation.

As a result, the network can learn:

* Curves
* Complex decision boundaries
* Feature interactions
* Hierarchical representations

---

## Intuition

### Without Activation

Each layer only scales and mixes numbers.

Stacking many layers:

```text
Linear → Linear → Linear
```

still behaves like:

```text
One Linear Layer
```

### With Activation

Each layer can make decisions:

```text
Activate neuron?
Yes / No
```

Different inputs activate different neurons.

This allows complex behavior to emerge.

---

## ReLU (Most Important to Understand)

Formula:

[
ReLU(x)=\max(0,x)
]

Behavior:

```python
if x < 0:
    return 0
else:
    return x
```

Advantages:

* Fast
* Simple
* Cheap to compute
* Helps gradient flow
* Works well in deep networks

Limitation:

* Dying ReLU problem (some neurons can become permanently inactive)

---

## Why Deep Learning Works

Activation functions allow the network to approximate complex functions.

Without activations:

```text
Only straight lines
```

With activations:

```text
Curves
Shapes
Complex boundaries
Almost any continuous function
```

This idea is related to the Universal Approximation Theorem.

---

# Common Activation Functions in Industry (2026)

## 1. ReLU

Used in:

* CNNs
* Vision models
* Mobile models
* Traditional deep learning systems

Formula:

[
ReLU(x)=\max(0,x)
]

---

## 2. GELU

Used in:

* BERT
* GPT-style models
* Transformers
* Modern NLP systems

Intuition:

Instead of a hard on/off decision like ReLU, GELU smoothly decides how much information to pass.

Advantages:

* Smoother gradients
* Better optimization
* Works well in very deep Transformers

---

## 3. SwiGLU

Used in:

* Llama family
* Modern LLMs
* State-of-the-art Transformer feed-forward networks

Advantages:

* Better parameter efficiency
* Better scaling behavior
* Often improves model quality

Important for modern LLM architecture.

---

## 4. Sigmoid

Formula:

[
\sigma(x)=\frac{1}{1+e^{-x}}
]

Output Range:

```text
0 to 1
```

Common Use:

* Binary classification output layer

Examples:

* Spam / Not Spam
* Fraud / Not Fraud
* Disease / No Disease

Rarely used in hidden layers today because of vanishing gradients.

---

## 5. Softmax

Used in:

* Multi-class classification output layers

Converts logits into probabilities.

Example:

```text
Cat   → 5%
Dog   → 90%
Horse → 3%
Bird  → 2%
```

Probabilities always sum to 100%.

# Interview Takeaways

### Why can't we remove activation functions?

Because multiple linear layers collapse into one linear layer.

The network would lose the ability to learn complex patterns.

---

### Why isn't sigmoid used everywhere?

Because:

* Saturates easily
* Causes vanishing gradients
* Makes deep networks difficult to train

---

### Why do modern Transformers prefer GELU/SwiGLU over ReLU?

Because:

* Smoother optimization
* Better gradient flow
* Better empirical performance in large-scale language models

## Exploding Gradient
**Gradients become extremely large during backpropagation**
**Causes unstable training and NaN losses**
**Fixed using gradient clipping, good initialization, normalization, and adaptive optimizers**


# Weight Initialization Cheat Sheet

## Why do we need it?
Goal:
- Prevent **vanishing gradients**
- Prevent **exploding gradients**
- Keep activations and gradients stable across layers

---

## Xavier (Glorot) Initialization
**Best for:** Sigmoid, Tanh

### Xavier Normal
\[
W \sim N\left(0,\frac{2}{fan_{in}+fan_{out}}\right)
\]

### Xavier Uniform
\[
W \sim U\left(-\sqrt{\frac{6}{fan_{in}+fan_{out}},\sqrt{\frac{6}{fan_{in}+fan_{out}}}\right)
\]

**Intuition:** Balances forward and backward signal flow.

---

## He (Kaiming) Initialization
**Best for:** ReLU, Leaky ReLU

### He Normal
\[
W \sim N\left(0,\frac{2}{fan_{in}}\right)
\]

### He Uniform
\[
W \sim U\left(-\sqrt{\frac{6}{fan_{in}}},\sqrt{\frac{6}{fan_{in}}}\right)
\]

**Intuition:** ReLU kills ~50% of activations, so larger initial weights are needed.

---

## Normal vs Uniform
- **Normal:** Bell-shaped distribution
- **Uniform:** All values equally likely

Usually:
> Xavier vs He matters much more than Normal vs Uniform.

---

## Interview Rule of Thumb

| Activation | Initialization |
|------------|---------------|
| Sigmoid | Xavier |
| Tanh | Xavier |
| ReLU | He/Kaiming |
| Leaky ReLU | He/Kaiming |

### Most Common Combination
```python
nn.ReLU()
nn.init.kaiming_normal_()
```
