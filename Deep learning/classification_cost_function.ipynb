# Classification Loss Functions Cheat Sheet

## 1. Binary Cross Entropy (BCE)
**Use:** Binary Classification (2 classes)

**Examples:**
- Spam / Not Spam
- Fraud / Not Fraud
- Disease / No Disease

**Output:** Sigmoid (1 neuron)

**Industry:** Gmail spam detection, fraud detection

---

## 2. Categorical Cross Entropy (CCE)
**Use:** Multi-Class Classification (exactly one correct class)

**Examples:**
- Cat, Dog, Horse
- Digit Recognition (0–9)

**Labels:** One-Hot Encoded

```python
[0, 1, 0]
```

**Output:** Softmax

**Industry:** Image classification, LLM next-token prediction

---

## 3. Sparse Categorical Cross Entropy (SCCE)
**Use:** Same as CCE, but labels are integers

```python
1
```

instead of

```python
[0, 1, 0]
```

**Output:** Softmax

**Industry:** Large-scale datasets like ImageNet

**Note:** Same loss as CCE; only label format differs.

---

## 4. Multi-Label Classification
**Use:** Multiple classes can be correct simultaneously

**Examples:**
- Movie genres
- Multiple diseases
- Image tagging

**Output:** Sigmoid per class

**Loss:** Binary Cross Entropy (BCE)

**Industry:** Netflix tagging, medical diagnosis

---

# Quick Rule

| Problem Type | Activation | Loss |
|-------------|------------|------|
| Binary Classification | Sigmoid | BCE |
| Multi-Class (One Correct Class) | Softmax | CCE |
| Multi-Class (Integer Labels) | Softmax | SCCE |
| Multi-Label | Sigmoid | BCE |

### Memory Trick
- **BCE** → "Is this class present?"
- **CCE** → "Which single class is correct?"
- **SCCE** → CCE + integer labels
- **Softmax** → Classes compete
- **Sigmoid** → Classes independent
