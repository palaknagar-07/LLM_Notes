

## 4. Why Same Embedding Model Is Required

Similarity search only works when:

```text
Document Embeddings
AND
Query Embeddings
```

come from the same embedding space.

### Wrong

```text
Documents → Model A
Queries   → Model B
```

### Correct

```text
Documents → Model A
Queries   → Model A
```

---

## 5. Dimension Size Is NOT The Real Problem

Even if:

```text
Model A → 15 dimensions
Model B → 15 dimensions
```

they may still produce incompatible embeddings.

Reason:

* Different training
* Different architecture
* Different learned weights
* Different semantic space

The issue is not dimensions.

The issue is different coordinate systems.

---

## 6. Embedding Models Create A Semantic Space

Think of an embedding model as creating its own universe.

Inside that universe:

```text
Dog  ↔ Cat
Paris ↔ France
Python ↔ Programming
```

are close together.

Two different embedding models create two different semantic universes.

---

## 7. Embedding Generation Pipeline

### NOT

```text
Text
 ↓
Token Vectors
 ↓
Average
 ↓
Embedding
```

### ACTUAL PROCESS

```text
Text
 ↓
Tokenizer
 ↓
Token IDs
 ↓
Initial Token Embeddings
 ↓
Transformer Layers
 ↓
Contextual Token Embeddings
 ↓
Pooling
 ↓
Final Embedding Vector
```

---

## 8. Contextual Embeddings

Initial token vectors are NOT the final vectors.

Example:

```text
I deposited money in the bank.
```

vs

```text
I sat near the river bank.
```

The word:

```text
bank
```

starts with the same initial embedding.

After self-attention:

```text
bank(finance)
```

and

```text
bank(river)
```

become different contextual vectors.

---

## 9. Transformer Runs During Embedding Generation

When calling:

```python
embed_query(...)
embed_documents(...)
```

the transformer actually runs.

It is NOT a simple lookup table.

The model performs:

* Self-attention
* Matrix multiplications
* Context understanding
* Pooling

to create the final embedding.

---

## 10. No Training Happens During Embedding Generation

When generating embeddings:

```text
✅ Forward Pass
✅ Inference

❌ Backpropagation
❌ Gradient Updates
❌ Weight Updates
```

The model is thinking, not learning.

---

## 11. Why PyTorch Is Needed

PyTorch is used to execute:

```text
Token Embeddings
 ↓
Attention
 ↓
Transformer Layers
 ↓
Pooling
```

efficiently.

It is not updating weights during embedding generation.

---

## 12. GPU Requirement

For RAG projects like RFQ processing:

```text
PDF
 ↓
Chunking
 ↓
Embeddings
 ↓
Vector DB
```

GPU is helpful but NOT required.

CPU works perfectly fine for:

* Nomic Embed
* BGE
* MiniLM
* E5

GPU mainly improves speed.

---

## One-Line Summary

**An embedding model does not simply assign stored vectors to words. It runs a transformer to create context-aware representations, pools them into a single embedding, and stores that embedding alongside the original text in a vector database for similarity search.**
