# System Arguments (`sys.argv`) - Revision

## Core Idea

A command-line program is like a function that receives arguments when it starts.

### Normal Function

```python
def add(a, b):
    return a + b

add(2, 3)
```

Arguments are accessed through parameter names (`a`, `b`).

### Command-Line Program

```bash
python factorial.py 7
```

The program receives arguments when it starts.

Python automatically creates:

```python
sys.argv = [
    "factorial.py",  # index 0
    "7"              # index 1
]
```

Access them using:

```python
import sys

n = int(sys.argv[1])
```

## Flow

```text
n = 7
        ↓
f"python factorial.py {n}"
        ↓
"python factorial.py 7"
        ↓
ShellTool executes command
        ↓
Python creates sys.argv
        ↓
["factorial.py", "7"]
        ↓
sys.argv[1] → "7"
        ↓
int("7") → 7
        ↓
factorial(7)
```

## Key Takeaway

* Functions receive arguments via parameters.
* Programs receive arguments via the command line.
* Python exposes command-line arguments through `sys.argv`.
* `sys.argv[0]` is always the script name.
* User-provided arguments start from `sys.argv[1]`.
