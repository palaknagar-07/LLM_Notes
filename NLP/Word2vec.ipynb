# Word2Vec — Simple Summary

---

# 1. What is Word2Vec?

Word2Vec is a method to convert words into vectors (numbers) so that:

> Similar words get similar vectors.

Before Word2Vec:
- Computers treated every word as completely unrelated.

Example:
```python
"cat" and "dog"
```

were as unrelated as:
```python
"cat" and "table"
```

Word2Vec changed this.

---

# 2. How Do We Decide the Feature Space?

We DON'T manually define features.

The model learns them automatically.

You simply provide:
- millions of sentences

and the model discovers patterns itself.

---

## Example

A word vector may internally capture concepts like:
- animateness
- gender
- royalty
- size
- emotions

But:

> These features have NO names.

They are just numbers.

---

## Real Word2Vec

Typically:
```python
300-dimensional vectors
```

Meaning:
```python
"cat" → [0.21, -0.44, 0.91, ... 300 values]
```

Each number captures some hidden property learned from language.

---

# 3. How Does Word2Vec Learn?

---

## Step 1 — Initialize Random Vectors

Initially every word gets random meaningless numbers.

Example:
```python
"cat" → [0.12, -0.89, 0.44]
"dog" → [-0.55, 0.33, -0.91]
```

At this point:
- vectors contain no meaning

---

## Step 2 — Read Sentences

The model scans text continuously.

Example sentence:
```python
"the cat sat on the mat"
```

For the word:
```python
"sat"
```

neighbor/context words are:
```python
["the", "cat", "on", "the"]
```

---

## Step 3 — Predict Neighbor Words

This is the most important idea.

The model asks:

> "Using the vector of 'sat', can I correctly guess nearby words?"

Example:
```python
Input word: "sat"

Expected nearby words:
cat, on, the
```

If prediction is wrong:
- slightly adjust vector values

Repeat:
- billions of times

---

# The Important Insight

Word2Vec is NOT directly taught meaning.

Instead:

> Meaning emerges automatically while solving the prediction task.

---

# Why Predicting Neighbors Captures Meaning

Words used in similar situations develop similar vectors.

Example:

## "sat" appears near:
```python
cat, dog, chair, floor
```

## "ate" appears near:
```python
food, pizza, hungry
```

## "barked" appears near:
```python
dog, loudly, owner
```

To predict these neighbors correctly:
- the vectors must encode semantic relationships

So eventually:
```python
cat ≈ dog
king ≈ queen
apple ≈ mango
```

because they appear in similar contexts.

---

# Core Principle

## Distributional Hypothesis

> "Words appearing in similar contexts tend to have similar meanings."

This is the entire foundation of Word2Vec.

# 4. Two Main Architectures

---

## CBOW (Continuous Bag of Words)

### Task:
Predict center word from surrounding context.

Example:
```python
Context: "the ___ sat on"

Predict:
"cat"
```

Fast and stable.

---

## Skip-Gram

### Task:
Predict surrounding words from center word.

Example:
```python
Input:
"cat"

Predict:
sat, the, on
```

Better for rare words.

Most famous Word2Vec implementations use Skip-Gram.

---

# 5. What Does the Vector Learn?

The amazing part:

Vectors learn semantic structure.

Example:
```python
king - man + woman ≈ queen
```

This shocked researchers when discovered.

It means:
- vectors encode relationships geometrically

# 6. Problems with Word2Vec

---

## Problem 1 — One Word = One Vector

Word2Vec gives ONE fixed vector per word.

Example:
```python
bank
```

has same vector in:
- river bank
- savings bank

This is wrong.

---

## Problem 2 — No Context Awareness

Word meaning changes by sentence.

Word2Vec cannot adapt dynamically.

---

## Problem 3 — OOV (Out Of Vocabulary)

Unseen words cannot be represented.

Example:
```python
"ChatGPT"
```

if not in training data → unknown.

---

# 7. What Replaced Word2Vec?

Transformers.

Especially:
- BERT
- Sentence Transformers

---

## Why They Are Better

They generate:
> Contextual embeddings

Meaning:
```python
"bank" in river sentence
```

gets different vector than:
```python
"bank" in finance sentence
```

Huge improvement.

## Both problems — same root cause
Word2Vec treats every word as a whole unit.
"cat" → one block. Either you know it fully or you don't know it at all.
It never looks inside the word. It doesn't know that:

"serendipitous" and "serendipity" are related
"unhappy" contains "happy"
"playing", "played", "plays" all come from "play"


How was this fixed? — FastText
Facebook made a small but powerful change.
Instead of treating the whole word as one unit, break the word into small character chunks.

"eating" → ea, eat, ati, tin, ing, eating

These chunks are called character n-grams.
Now the vector for "eating" is built from the vectors of all its chunks.
Why does this help?

"eating", "eats", "eaten" — all share the chunk "eat" — so they automatically get similar vectors
New word "ChatGPT" — even if never seen — its character chunks might have been seen in other words — so it gets a rough but real vector instead of nothing
