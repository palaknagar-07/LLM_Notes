# The Full Classical NLP Pipeline — How Models Deal with Vectors

## 1. Text Preprocessing

You start with raw text.

Before any model sees it, you preprocess it:

- Convert text to lowercase
- Tokenize into words
- Remove stopwords (`the`, `a`, `is`, etc.) to reduce vocabulary size
- Apply stemming or lemmatization to merge word variants

Examples:

- `running → run`
- `mangoes → mango`

The goal is to ensure that the same concept does not occupy multiple dimensions.

---

## 2. Building the Vocabulary

After preprocessing, you build a shared vocabulary across the entire corpus.

Every unique word receives a fixed index.

Example:

| Word | Index |
|--------|--------|
| apple | 0 |
| mango | 1 |
| run | 2 |
| machine | 3 |

If the corpus contains **50,000 unique words**, then every document will be represented by a vector containing exactly **50,000 dimensions**.

---

## 3. Converting Documents into Vectors

### Bag of Words (BoW)

Each dimension stores the raw count of a word.

Example:

Sentence:

> I love machine learning machine learning

Vector:

```
[I:1, love:1, machine:2, learning:2, ...]
```

### TF-IDF

Instead of raw counts, each dimension stores a weighted score:

- High if the word appears frequently in this document
- Low if the word appears in many documents across the corpus

This reduces the importance of common words and highlights informative words.

---

## 4. Documents Become Points in High-Dimensional Space

The resulting vector is simply a point in a 50,000-dimensional space.

The model never reads words directly.

It only sees:

```
[0, 3, 0, 5, 0, 0, 2, ...]
```

Everything is numbers.

---

## 5. Feeding Vectors into a Model

These vectors are provided to a machine learning model along with labels.

Examples:

| Document | Label |
|-----------|--------|
| "Win cash now" | Spam |
| "Meeting at 3 PM" | Not Spam |

or

| Review | Label |
|---------|---------|
| "Amazing movie" | Positive |
| "Terrible experience" | Negative |

Training means learning how to map input vectors to output labels.

---

# Logistic Regression

## Parameters

For a vocabulary of 50,000 words:

- 50,000 weights
- 1 bias term

Each word gets its own learned weight.

Example:

| Word | Weight |
|--------|--------|
| prize | +4.2 |
| winner | +3.8 |
| friend | -2.1 |
| meeting | -1.7 |

The model learns which words are useful for predicting a class.

---

## Prediction

For a new document:

1. Take its vector
2. Multiply each dimension by its corresponding weight
3. Sum everything
4. Add bias
5. Pass through a sigmoid function

Mathematically:

```
z = w₁x₁ + w₂x₂ + ... + wₙxₙ + b
```

Then:

```
Probability = Sigmoid(z)
```

The sigmoid squashes the result into a value between 0 and 1.

Words absent from the document contribute zero because their counts are zero.

---

# Support Vector Machine (SVM)

## Parameters

Instead of learning probabilities, SVM learns a **hyperplane**.

A hyperplane is simply a decision boundary in high-dimensional space.

Example:

```
Spam          | Hyperplane | Not Spam
```

The goal is to find the boundary that maximizes the margin between classes.

---

## Support Vectors

The training points closest to the boundary are called:

**Support Vectors**

These points determine where the hyperplane sits.

All other points are less important.

---

## Prediction

For a new document vector:

- Determine which side of the hyperplane it falls on
- Assign the corresponding class

---

# How Training Works

Both Logistic Regression and many other machine learning models learn using **Gradient Descent**.

The process:

1. Make a prediction
2. Compare prediction with true label
3. Compute error
4. Adjust weights slightly to reduce error
5. Repeat over thousands of examples

Over time:

- Useful words receive stronger weights
- Unhelpful words receive weaker weights

Training continues until performance stops improving.

---

# The Fundamental Limitation of Classical NLP

Every word receives exactly **one fixed representation**.

Example:

```
bank
```

appears in:

> I deposited money in the bank.

and

> We sat beside the river bank.

Both sentences use the same vocabulary index:

```
bank → 1542
```

The model cannot distinguish between the two meanings.

Context has already been lost before the model sees the input.

---

# Why Embeddings and Transformers Were Created

Classical approaches treat words as isolated symbols.

Modern approaches learn contextual representations.

Instead of:

```
bank → one fixed vector
```

they learn:

```
bank (finance) → vector A
bank (river)   → vector B
```

depending on surrounding words.

This solves one of the biggest weaknesses of Bag of Words and TF-IDF.

---

# The Key Takeaway

The overall pipeline remains surprisingly similar:

1. Convert text into vectors
2. Feed vectors into a model
3. Compute a score
4. Compare with the true label
5. Adjust parameters using gradient descent

The major difference is that modern NLP uses far richer vector representations, allowing models to capture meaning and context rather than just word counts.
